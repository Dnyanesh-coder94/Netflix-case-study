# *Unnesting Object Columns*

```python
def un_nest(df, column):
    df[column] = df[column].str.split(', ')
    df = df.explode(column)
    df[column] = df[column].str.strip()
    return df
# Columns to un-nest
columns_to_unnest = ['director', 'cast', 'country', 'listed_in']
# Un-nesting selected columns
for column in columns_to_unnest:
    df = un_nest(df, column)

df
```

```python
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

# Detect outliers in the 'release_year' column
outliers_release_year = detect_outliers_iqr(df['release_year'])
print("Outliers in 'release_year' column:\n", df[outliers_release_year]['release_year'])
```
